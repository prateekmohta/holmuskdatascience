```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading CSVs into dataframes
bill_amount = pd.read_csv('bill_amount.csv')
bill_id = pd.read_csv('bill_id.csv')
clinical_data = pd.read_csv('clinical_data.csv')
demographics = pd.read_csv('demographics.csv')
demographics
```

```python
#joining CSVs to create a consolidated dataset
merge = pd.merge(bill_id,bill_amount,how = 'left',on='bill_id')
merge1 = pd.merge(merge,demographics,how='left',on='patient_id')
merge2 = pd.merge(merge1,clinical_data,how='left',on=['patient_id','date_of_admission'])
```

```python
merge2
```

```python
#Identify columns having missing values
merge2.columns[merge2.isna().any()].tolist()
```

```python
#Cleaning data by renaming categories in following categorical variables
merge2['gender'] = merge2['gender'].replace({'m': 'Male', 'f': 'Female'})
merge2['race'] = merge2['race'].replace({'India':'Indian','chinese':'Chinese'})
merge2['resident_status'] = merge2['resident_status'].replace({'Singapore citizen':'Singaporean'})
merge2['medical_history_3'] = merge2['medical_history_3'].replace({'No':'0','Yes':'1'})

#Assigning a class to missing values
merge2['medical_history_2'] = merge2.medical_history_2.fillna('Not Specified')
merge2['medical_history_5'] = merge2.medical_history_5.fillna('Not Specified')
```

```python
#Verifiying that there are no missing values
merge2.isnull().values.any()
```

```python
#Feature Engineering

merge2['date_of_admission'] = pd.to_datetime(merge2['date_of_admission'])
merge2['date_of_discharge'] = pd.to_datetime(merge2['date_of_discharge'])
merge2['date_of_birth'] = pd.to_datetime(merge2['date_of_birth'])

merge2['number_of_days'] = np.round((merge2['date_of_discharge'] - merge2['date_of_admission'])/np.timedelta64(1, 'D')).astype(int) 
merge2['age_at_admission'] = np.round(((merge2['date_of_admission'] - merge2['date_of_birth'])/ np.timedelta64(1, 'D'))/365.25).astype(int)
merge2['bmi'] = merge2['weight']/(merge2['height']/100)**2
merge2['weight_classification'] = pd.cut(merge2['bmi'],bins=[0,18.5,24.9,29.9,39.9,50],labels=['underweight','normal','overweight','obese','extremely obese'])
```
